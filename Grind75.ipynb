```python
#2 Sum 
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                sum = nums[i] + nums[j]
                if target  == sum:
                    return(i,j)

#Intuition Approach
#TC - O(n^2)
#SC - O(1)

#Optimzed Approach
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        seen = {}
        for i in range(len(nums)):
            rem = target - nums[i]
            if rem in seen:
                return (i,seen[rem])
            seen[nums[i]] = i
            
#Classic problem of space and time tradeof            
#TC - O(n)
#SC - O(n)
```

```python
#Valid Parenthesis
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        parentesis = {')':'(','}':'{',']':'['}
        for ch in s:
            if ch not in parentesis:
                stack.append(ch)
            else:
                if not stack:
                    return False
                top = stack.pop()
                if top != parentesis[ch]:
                    return False

        if len(stack) == 0:
            return True
        else:
            return False
        
#Time Complexity  : O(n) -for loop and all stack opn are O(1)
#Space Complexity : O(n)  - stack -O(n) and haspmap is fixed only 3 so O(1) a

        
```

```python
#121. Best Time to Buy and Sell Stock
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        min_price = prices[0]
        profit = 0
        for i in range(len(prices)):
            if prices[i] < min_price :
                min_price = prices[i]
            profit = max(profit,prices[i] - min_price)
        return profit

#TC - O(n)
#SC - O(1)
```

```python
#242. Valid Anagram
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        freqCounts,freqCountt = {},{}
        for i in range(len(s)):
            freqCounts[s[i]] = freqCounts.get(s[i],0) + 1
            freqCountt[t[i]] = freqCountt.get(t[i],0) + 1

        if freqCounts == freqCountt:
            return True
        else:
            return False
#TC - O(n)
#Sc - O(1) -> since alphabets are limited to 26
```

```python
#704 Binary Search
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        left = 0
        right = len(nums)-1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                return mid
            if nums[mid] <= target:
                left = mid + 1
            else:
                right = mid - 1
        return -1
#TC - O(logn)
#SC - O(1)
```

```python
#977. Squares of a Sorted Array
class Solution(object):
    def sortedSquares(self, nums):
        i = 0
        j = len(nums)-1
        pos = len(nums) 
        sortedarray = [0] * pos
        print("sorted array:",sortedarray,"pos:",pos,"j:",j)
        while i <= j:
            lsq = nums[i] * nums[i]
            rsq = nums[j] * nums[j]
            if lsq > rsq:
                sortedarray[pos-1] = lsq
                i += 1
            else:
                sortedarray[pos-1] = rsq
                j -= 1
            pos -= 1
            
        return sortedarray
        
#TC - O(n)
#SC - O(n)
```

```python
#GRAPH BFS
from collections import deque

def bfs(start, graph, n):
    visited = [False] * n          # visited array
    queue = deque()                # queue for BFS

    queue.append(start)
    visited[start] = True

    while queue:
        node = queue.popleft()
        print(node, end=" ")

        for neighbor in graph[node]:
            if not visited[neighbor]:
                visited[neighbor] = True
                queue.append(neighbor)
#TC - O(V + E)
#SC - O(v)
```

```python
#flood Fill
class Solution(object):
    def floodFill(self, image, sr, sc, color):
        originalColor = image[sr][sc]

        # edge case: avoid infinite loop
        if originalColor == color:
            return image

        rows, cols = len(image), len(image[0])

        def dfs(r, c):
            # 1. base cases (STOP)
            if r < 0 or c < 0 or r >= rows or c >= cols:
                return
            if image[r][c] != originalColor:
                return

            # 2. do work (mark visited)
            image[r][c] = color

            # 3. go deep (4 directions)
            dfs(r + 1, c)  # down
            dfs(r - 1, c)  # up
            dfs(r, c + 1)  # right
            dfs(r, c - 1)  # left

        dfs(sr, sc)
        return image
```

```python
#226. Invert Binary Tree - BFS Approach
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        queue = deque([root])
        if not root:
            return
        while queue:
            node = queue.popleft()
            node.left,node.right = node.right,node.left  
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        return root
#TC - O(n) - since we are visiting all node
#SC - O(n) queue
#226. Invert Binary Tree - DFS Approach
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        if not root:
            return
        root.left,root.right = root.right,root.left  
        self.invertTree(root.left)
        self.invertTree(root.right)
        return root
#TC - O(n) - since we are visiting all node
#SC - O(h) - stack, for skewed tree -O(n), balanced tree -O(logn)
```

```python
#543 diamter of binary tree
class Solution:
    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        res = 0
        def dfs(node):
            if not node:
                return 0
            lh = dfs(node.left)
            rh = dfs(node.right)
            nonlocal res
            res = max(res , lh + rh)       
            return 1 + max(lh,rh)
        dfs(root)
        return res
#TC - O(n)
#SC - O(h) , worst case -O(n)
```

```python
#21 merge two sorted lists
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        curr1 = list1
        curr2 = list2
        newlist = ListNode(0)
        curr3 = newlist
        while curr1 and curr2:
            if curr1.val < curr2.val:
                curr3.next = curr1
                curr1 = curr1.next
            else:
                curr3.next = curr2
                curr2 = curr2.next
            curr3 = curr3.next
        curr3.next = curr1 or curr2
        return newlist.next

#TC-O(n1+n2)
#SC - O(1)  because we are resuing the existing nodes
         
```

```python
#206 Reverse Linked List
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        prev = None
        if not head or not head.next:
            return head
        curr = head      
        while curr:
            front = curr.next
            curr.next = prev
            prev = curr
            curr = front
        return prev
#TC - O(n)
#SC - O(1)
        
```
