```python
#2 Sum 
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                sum = nums[i] + nums[j]
                if target  == sum:
                    return(i,j)

#Intuition Approach
#TC - O(n^2)
#SC - O(1)

#Optimzed Approach
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        seen = {}
        for i in range(len(nums)):
            rem = target - nums[i]
            if rem in seen:
                return (i,seen[rem])
            seen[nums[i]] = i
            
#Classic problem of space and time tradeof            
#TC - O(n)
#SC - O(n)
```

```python
#Valid Parenthesis
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        parentesis = {')':'(','}':'{',']':'['}
        for ch in s:
            if ch not in parentesis:
                stack.append(ch)
            else:
                if not stack:
                    return False
                top = stack.pop()
                if top != parentesis[ch]:
                    return False

        if len(stack) == 0:
            return True
        else:
            return False
        
#Time Complexity  : O(n) -for loop and all stack opn are O(1)
#Space Complexity : O(n)  - stack -O(n) and haspmap is fixed only 3 so O(1) a

        
```

```python
#121. Best Time to Buy and Sell Stock
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        min_price = prices[0]
        profit = 0
        for i in range(len(prices)):
            if prices[i] < min_price :
                min_price = prices[i]
            profit = max(profit,prices[i] - min_price)
        return profit

#TC - O(n)
#SC - O(1)
```

```python
#217. Contains Duplicate using Hashmap
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        countholder = {}
        for i in range(len(nums)):
            countholder[nums[i]] = countholder.get(nums[i],0) + 1
            if countholder.get(nums[i]) > 1:
                return True
        return False

#TC - O(n)
#SC - O(n)

# Using HashSet          
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        seen = set()
        for num in nums:
            if num in seen:
                return True
            seen.add(num)
        return False
#TC - O(n)
#SC - O(n)
```

```python
#242. Valid Anagram
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        freqCounts,freqCountt = {},{}
        for i in range(len(s)):
            freqCounts[s[i]] = freqCounts.get(s[i],0) + 1
            freqCountt[t[i]] = freqCountt.get(t[i],0) + 1

        if freqCounts == freqCountt:
            return True
        else:
            return False
#TC - O(n)
#Sc - O(1) -> since alphabets are limited to 26
```

```python
#409. Longest Palindrom
class Solution:
    def longestPalindrome(self, s: str) -> int:
        frqcount,output = {},0
        hasOdd = False
        for ch in s:
            frqcount[ch] = frqcount.get(ch,0) + 1
        for count in frqcount.values():
            output += (count//2) * 2 #pairs
            if count % 2 == 1:
                hasOdd = True
            
        if hasOdd:
            output += 1
        return output
#TC - O(n)
#SC - O(1)


#Use all pairs from every character, then add one center if any odd exists.
```

```python
#704 Binary Search
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        left = 0
        right = len(nums)-1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                return mid
            if nums[mid] <= target:
                left = mid + 1
            else:
                right = mid - 1
        return -1
#TC - O(logn)
#SC - O(1)
```

```python
#977. Squares of a Sorted Array
class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        l = 0
        r = len(nums) - 1
        pos = len(nums) -1
        sortedarray = [0] * len(nums)
        while l <= r:
            lsq = nums[l] * nums[l]
            rsq = nums[r] * nums[r]

            if lsq > rsq:
                sortedarray[pos] = lsq
                l += 1
            else:
                sortedarray[pos] = rsq
                r -= 1
            pos -=1
        return sortedarray
#TC - O(n)
#SC - O(n)
```

```python
#1752 check is array is sorted if needed rotate
class Solution:
    def check(self, nums: List[int]) -> bool:
        n,count = len(nums),0
        for i in range(n):
            if nums[i] > nums[(i+1)%n]:
                count += 1
                if count > 1:
                    return False
        return True
#TC - O(n)
#sc -O(1)
            
```

```python
#26 Remove Duplicates from Sorted Array
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        i = 0
        for j in range(len(nums)):
            if nums[i] != nums[j]:
                i += 1
                nums[i] = nums[j]
        return i+1
#TC O(n)
#SC O(1)
```

```python
#1480 Running sum of 1 d array
class Solution(object):
    def runningSum(self, nums):
        for i in range(1,len(nums)):
            nums[i] = nums[i-1] + nums[i]
        return nums
#TC- O(n)
#sc - O(1)
```

```python
#1672. Richesr Customer Wealth
class Solution(object):
    def maximumWealth(self, accounts):
        maxsum = 0
        for i in accounts:
            curr_sum  = 0 
            for j in i:
                curr_sum  += j
            maxsum = max(maxsum,curr_sum )
        return maxsum
#TC - O(m*n)
#SC - O(1)
```
