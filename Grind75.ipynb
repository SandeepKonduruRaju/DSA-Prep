```python
#2 Sum 
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                sum = nums[i] + nums[j]
                if target  == sum:
                    return(i,j)

#Intuition Approach
#TC - O(n^2)
#SC - O(1)

#Optimzed Approach
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        seen = {}
        for i in range(len(nums)):
            rem = target - nums[i]
            if rem in seen:
                return (i,seen[rem])
            seen[nums[i]] = i
            
#Classic problem of space and time tradeof            
#TC - O(n)
#SC - O(n)
```

```python
#Valid Parenthesis
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        parentesis = {')':'(','}':'{',']':'['}
        for ch in s:
            if ch not in parentesis:
                stack.append(ch)
            else:
                if not stack:
                    return False
                top = stack.pop()
                if top != parentesis[ch]:
                    return False

        if len(stack) == 0:
            return True
        else:
            return False
        
#Time Complexity  : O(n) -for loop and all stack opn are O(1)
#Space Complexity : O(n)  - stack -O(n) and haspmap is fixed only 3 so O(1) a

        
```
