```python
#2 Sum 
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                sum = nums[i] + nums[j]
                if target  == sum:
                    return(i,j)

#Intuition Approach
#TC - O(n^2)
#SC - O(1)

#Optimzed Approach
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        seen = {}
        for i in range(len(nums)):
            rem = target - nums[i]
            if rem in seen:
                return (i,seen[rem])
            seen[nums[i]] = i
            
#Classic problem of space and time tradeof            
#TC - O(n)
#SC - O(n)
```

```python
#Valid Parenthesis
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        parentesis = {')':'(','}':'{',']':'['}
        for ch in s:
            if ch not in parentesis:
                stack.append(ch)
            else:
                if not stack:
                    return False
                top = stack.pop()
                if top != parentesis[ch]:
                    return False

        if len(stack) == 0:
            return True
        else:
            return False
        
#Time Complexity  : O(n) -for loop and all stack opn are O(1)
#Space Complexity : O(n)  - stack -O(n) and haspmap is fixed only 3 so O(1) a

        
```

```python
#121. Best Time to Buy and Sell Stock
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        min_price = prices[0]
        profit = 0
        for i in range(len(prices)):
            if prices[i] < min_price :
                min_price = prices[i]
            profit = max(profit,prices[i] - min_price)
        return profit

#TC - O(n)
#SC - O(1)
```

```python
#217. Contains Duplicate using Hashmap
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        countholder = {}
        for i in range(len(nums)):
            countholder[nums[i]] = countholder.get(nums[i],0) + 1
            if countholder.get(nums[i]) > 1:
                return True
        return False

#TC - O(n)
#SC - O(n)

# Using HashSet          
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        seen = set()
        for num in nums:
            if num in seen:
                return True
            seen.add(num)
        return False
#TC - O(n)
#SC - O(n)
```

```python
#125. Valid Palindrome
class Solution:
    def isPalindrome(self, s: str) -> bool:
        newString = ""
        for ch in s:
            if ch.isalnum():
                newString += ch.lower()
        return newString == newString[::-1] 
#TC - O(n)
#SC - O(n)

class Solution:
    def isPalindrome(self, s: str) -> bool:
        L = 0
        R = len(s) -1
        while L < R:
            if not s[L].isalnum():
                L = L+1
                continue
            if not s[R].isalnum():
                R = R-1
                continue
            print("S[L]",s[L].lower(),"S[R]",s[R].lower())
            if s[L].lower() == s[R].lower():
                L = L+1
                R = R-1
            else:
                return False
        
        return True
#TC - O(n)
#SC - O(1)
```

```python
#242. Valid Anagram
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        freqCounts,freqCountt = {},{}
        for i in range(len(s)):
            freqCounts[s[i]] = freqCounts.get(s[i],0) + 1
            freqCountt[t[i]] = freqCountt.get(t[i],0) + 1

        if freqCounts == freqCountt:
            return True
        else:
            return False
#TC - O(n)
#Sc - O(1) -> since alphabets are limited to 26
```

```python
#409. Longest Palindrom
class Solution:
    def longestPalindrome(self, s: str) -> int:
        frqcount,output = {},0
        hasOdd = False
        for ch in s:
            frqcount[ch] = frqcount.get(ch,0) + 1
        for count in frqcount.values():
            output += (count//2) * 2 #pairs
            if count % 2 == 1:
                hasOdd = True
            
        if hasOdd:
            output += 1
        return output
#TC - O(n)
#SC - O(1)


#Use all pairs from every character, then add one center if any odd exists.
```

```python
#704 Binary Search
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        left = 0
        right = len(nums)-1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                return mid
            if nums[mid] <= target:
                left = mid + 1
            else:
                right = mid - 1
        return -1
#TC - O(logn)
#SC - O(1)
```
