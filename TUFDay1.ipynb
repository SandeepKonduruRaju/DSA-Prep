```python
#73. Set Matrix Zeroes
# Brute Force Solution for Set Matrix Zeroes Problem
# Problem: Given an m x n matrix, if an element is 0, set its entire row and column to 0.

# Time Complexity: O((m * n) * (m + n))
# Space Complexity: O(1)

class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        # Get matrix dimensions
        rows = len(matrix)
        cols = len(matrix[0])

        # Temporary marker value for cells that need to become zero later
        X = -99999   # Use any unique value not present in the matrix

        # Step 1: Mark rows and columns that should become zero
        for i in range(rows):
            for j in range(cols):
                if matrix[i][j] == 0:
                    # Mark the entire row
                    for r in range(cols):
                        if matrix[i][r] != 0:
                            matrix[i][r] = X
                    # Mark the entire column
                    for c in range(rows):
                        if matrix[c][j] != 0:
                            matrix[c][j] = X

        # Step 2: Convert all marked cells to 0
        for i in range(rows):
            for j in range(cols):
                if matrix[i][j] == X:
                    matrix[i][j] = 0
```

```python
class Solution:
    def check(self, nums: List[int]) -> bool:
        n = len(nums)
        count = 0
        # Step 1: Find where order breaks (rotation point)
        for i in range(n - 1):
            if nums[i] > nums[i + 1]:
                count = (n - 1) - i   # how many times to rotate
                break   # once found, no need to check further
        
        # Step 2: Rotate 'count' times
        for i in range(count):
            last = nums[-1]  # store last element
            # shift all elements right
            for j in range(n - 1, 0, -1):
                nums[j] = nums[j - 1]
            nums[0] = last   # place last at first
        
        # Step 3: Check if array is sorted after rotation
        for i in range(n - 1):
            if nums[i] > nums[i + 1]:
                return False
            
        return True

#TC - O(n) + O(n*2) + O(n)
#SC - O(1) 

#Optimized Approch
class Solution:
    def check(self, nums: List[int]) -> bool:
        n = len(nums)
        drop_count = 0

        for i in range(n):
            if nums[i] > nums[(i + 1) % n]:
                drop_count =  drop_count + 1

        return drop_count <= 1

#TC - O(n)
#SC - O(1) 
```
