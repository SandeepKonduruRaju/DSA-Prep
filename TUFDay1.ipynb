```python
#73. Set Matrix Zeroes
# Brute Force Solution for Set Matrix Zeroes Problem
# Problem: Given an m x n matrix, if an element is 0, set its entire row and column to 0.

# Time Complexity: O((m * n) * (m + n))
# Space Complexity: O(1)

class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        # Get matrix dimensions
        rows = len(matrix)
        cols = len(matrix[0])

        # Temporary marker value for cells that need to become zero later
        X = -99999   # Use any unique value not present in the matrix

        # Step 1: Mark rows and columns that should become zero
        for i in range(rows):
            for j in range(cols):
                if matrix[i][j] == 0:
                    # Mark the entire row
                    for r in range(cols):
                        if matrix[i][r] != 0:
                            matrix[i][r] = X
                    # Mark the entire column
                    for c in range(rows):
                        if matrix[c][j] != 0:
                            matrix[c][j] = X

        # Step 2: Convert all marked cells to 0
        for i in range(rows):
            for j in range(cols):
                if matrix[i][j] == X:
                    matrix[i][j] = 0
```
