```python
#49. Group Anagrams
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        hashmap = {}
        for word in strs:
            key = tuple(sorted(word))
            if key not in hashmap:
                hashmap[key] = []
            hashmap[key].append(word)
        return list(hashmap.values())
#TC O(n * k log k)
#SC O(n * k)
```

```python
#347. Top K Frequent Elements.
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        hashmap = {}
        result = []
        for num in nums:
            hashmap[num] = hashmap.get(num,0) + 1

        for i in range(k):
            maxvalue = 0
            maxkey = None
            for key,value in hashmap.items():
                if value > maxvalue:
                    maxvalue = value
                    maxkey = key
            result.append(maxkey)
            hashmap[maxkey] = 0
        return result

#TC O(n) + O(n*K) =O(n*k)
#SC O(n+k) since k<=n -> O(n)


#More optimised solution using Sorted
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        hashmap = {}
        result = []
        for num in nums:
            hashmap[num] = hashmap.get(num,0) + 1

        sortedhash = sorted(hashmap.items(),key = lambda x: x[1],reverse = True)
        
        for i in range(k):
            result.append(sortedhash[i][0])

        return result    
#TC O(n + logn)
#SC O(n)
        
```

```python
#15. 3sum
class Solution:
    def threeSum(self, nums: list[int]) -> list[list[int]]:
        nums.sort()
        results = set()
        for i in range(len(nums)):
            if i > 0 and nums[i] == nums[i-1]:
                continue  # ✅ skip this duplicate i
            left = i + 1
            right = len(nums) - 1
            while left < right:
                sum = nums[i] + nums[left] + nums[right]
                if sum == 0:
                    results.add((nums[i],nums[left],nums[right]))
                    left = left + 1
                    right = right - 1
                elif sum < 0:
                    left = left + 1
                elif sum > 0:
                    right = right - 1
        return [list(t) for t in results]
    
#TC - O(n*2)
#SC - O(k) - k - no of triplets
```
