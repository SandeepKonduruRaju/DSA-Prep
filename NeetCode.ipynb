```python
#49. Group Anagrams
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        hashmap = {}
        for word in strs:
            key = tuple(sorted(word))
            if key not in hashmap:
                hashmap[key] = []
            hashmap[key].append(word)
        return list(hashmap.values())
#TC O(n * k log k)
#SC O(n * k)
```

```python
#347. Top K Frequent Elements.
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        hashmap = {}
        result = []
        for num in nums:
            hashmap[num] = hashmap.get(num,0) + 1

        for i in range(k):
            maxvalue = 0
            maxkey = None
            for key,value in hashmap.items():
                if value > maxvalue:
                    maxvalue = value
                    maxkey = key
            result.append(maxkey)
            hashmap[maxkey] = 0
        return result

#TC O(n) + O(n*K) =O(n*k)
#SC O(n+k) since k<=n -> O(n)


#More optimised solution using Sorted
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        hashmap = {}
        result = []
        for num in nums:
            hashmap[num] = hashmap.get(num,0) + 1

        sortedhash = sorted(hashmap.items(),key = lambda x: x[1],reverse = True)
        
        for i in range(k):
            result.append(sortedhash[i][0])

        return result    
#TC O(n + logn)
#SC O(n)
        
```

```python
#15. 3sum
class Solution:
    def threeSum(self, nums: list[int]) -> list[list[int]]:
        nums.sort()
        results = set()
        for i in range(len(nums)):
            if i > 0 and nums[i] == nums[i-1]:
                continue  # ✅ skip this duplicate i
            left = i + 1
            right = len(nums) - 1
            while left < right:
                sum = nums[i] + nums[left] + nums[right]
                if sum == 0:
                    results.add((nums[i],nums[left],nums[right]))
                    left = left + 1
                    right = right - 1
                elif sum < 0:
                    left = left + 1
                elif sum > 0:
                    right = right - 1
        return [list(t) for t in results]
    
#TC - O(n*2)
#SC - O(k) - k - no of triplets
```

```python
# 11. Container with Most water
class Solution:
    def maxArea(self, height: List[int]) -> int:
        L = 0
        maxwater = 0
        R = len(height) - 1
        while L < R:
            maxwater = max(maxwater, min(height[L],height[R]) * abs(L-R))
            if height[L] < height[R]:
                L = L + 1
            else:
                R = R - 1
        return maxwater
#TC - O(n)
#SC - O(1)
```

```python
#53 Maxsubarray
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        maxsum = nums[0]
        sum = 0
        for i in range(len(nums)):
            if sum < 0:
                sum = 0
            sum += nums[i]
            if sum > maxsum:
                maxsum = sum

        return maxsum
#TC -O(n)
#SC -O(1)
```

```python
#3. Longest Substring Without Repeating Characters
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        maxlen = 0
        for i in range(len(s)):
            seen = {}
            for j in range(i,len(s)):
                if s[j] in seen:
                    break
                seen[s[j]] = 1

                maxlen = max(maxlen,j-i+1)
        return maxlen
#TC - O(n^2)
#SC - O(n)

#optimising it using sliding window

# 3 Longest Substring w/o repeating ch
# 3 Longest Substring w/o repeating ch
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        maxlen = 0
        dict = {}
        L = 0
        for R in range(len(s)):
            if s[R] in dict and dict[s[R]] >= L:
                L = dict[s[R]]+1
            dict[s[R]] = R
            maxlen = max(maxlen,R-L+1)
        return maxlen
#TC O(n)
#SC O(n)
```
