```python
#15. 3sum
class Solution:
    def threeSum(self, nums: list[int]) -> list[list[int]]:
        nums.sort()
        results = set()
        for i in range(len(nums)):
            if i > 0 and nums[i] == nums[i-1]:
                continue  # ✅ skip this duplicate i
            left = i + 1
            right = len(nums) - 1
            while left < right:
                sum = nums[i] + nums[left] + nums[right]
                if sum == 0:
                    results.add((nums[i],nums[left],nums[right]))
                    left = left + 1
                    right = right - 1
                elif sum < 0:
                    left = left + 1
                elif sum > 0:
                    right = right - 1
        return [list(t) for t in results]
    
#TC - O(n*2)
#SC - O(k) - k - no of triplets
```

```python
# 11. Container with Most water
class Solution:
    def maxArea(self, height: List[int]) -> int:
        L = 0
        maxwater = 0
        R = len(height) - 1
        while L < R:
            maxwater = max(maxwater, min(height[L],height[R]) * abs(L-R))
            if height[L] < height[R]:
                L = L + 1
            else:
                R = R - 1
        return maxwater
#TC - O(n)
#SC - O(1)
```

```python
#39. Combination Sum
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        result = []

        def backtracking(i,target,path):
            if target == 0:
                result.append(path[:]) #create a copy IMP
                return
            if i > len(candidates)-1 or target < 0:
                return
            
            path.append(candidates[i])
            backtracking(i,target-candidates[i],path)
            path.pop()
            backtracking(i+1,target,path)

        backtracking(0,target,[])
        return result
#O(2^target)
#O(target)
```

```python
#153. Find Minimum in Rotated Sorted Array - better
class Solution:
    def findMin(self, nums: List[int]) -> int:
        l = 0 
        r = len(nums) - 1
        while l <= r:
            m = (l + r) // 2
            if l == r:        # ← check this FIRST!
                return nums[l]
            if nums[m] > nums[r]:
                l = m + 1
            else:
                r = m
#O(logn)
#O(1)
```

```python
#33. Search In Rotated Sorted Array
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        l = 0
        r = len(nums)-1
        while l <= r:
            m = (l + r) // 2
            if nums[m] == target:
                return m
            #either half is sorted when rotated
            if nums[l] <= nums[m]:
                #left half is sorted
                if nums[l] <= target <= nums[m]:
                    r = m - 1
                else:
                    l = m + 1
            else:
                #right half is sorted
                if nums[m] <= target <= nums[r]:
                    l = m + 1
                else:
                    r = m - 1
            
        return -1
#O(logn)
#O(1)
```

```python
#438. Find All Anagrams in a String
class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        plen = len(p)
        scount = {}
        result = []
        pcount = {}
        if len(s) < len(p):
            return [] 
        #PSTRING Count
        for ch in p:
            pcount[ch] = pcount.get(ch,0) + 1
        #S String Count
        for i in range(plen):
            scount[s[i]] = scount.get(s[i],0) + 1
        
        #add ch and delete a ch 
        for i in range(plen,len(s)):
            if pcount == scount:
                result.append(i-plen)
            #rgba(73, 95, 57, 1) char count
            scount[s[i]] = scount.get(s[i],0) + 1

            #delete left ch
            left = i - plen
            scount[s[left]] -= 1
            if scount[s[left]] == 0:
                del scount[s[left]]
        if pcount == scount:
                result.append(len(s)-plen)

        return result
#O(n)
#O(k))
```
