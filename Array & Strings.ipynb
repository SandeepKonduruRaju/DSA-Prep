```python
#57. Insert Interval
class Solution(object):
    def insert(self, intervals, newInterval):
        #traversing and appending a new ele
        for i in range(len(intervals)):
            if newInterval[0] <= intervals[i][0]:
                intervals.insert(i,newInterval)
                break
            else:
                intervals.append(newInterval)

        
        #optimise to overlap the intervals
        result = [intervals[0]]
        for i  in range(1,len(intervals)):
            prev = result[-1]
            curr = intervals[i]
            if curr[0] <= prev[1]:
                prev[1] = max(prev[1],curr[1])
            else:
                result.append(curr)

        return result
#TC = O(n)
#SC = O(n)
```

```python
class Solution(object):
    def insert(self, intervals, newInterval):
        result = []
        i = 0
        n = len(intervals)

        #insertion before overlap
        while i < n and intervals[i][1] < newInterval[0]:
            result.append(intervals[i])
            i+=1
        
        #insertion During overlap
        while i < n and intervals[i][0] <= newInterval[1]:
            newInterval[0] = min(newInterval[0],intervals[i][0])
            newInterval[1] = max(newInterval[1],intervals[i][1])
            i+=1
        result.append(newInterval)
            

        #insertion after overlap
        while i < n:
            result.append(intervals[i])
            i+=1
        
        return result
#TC = O(n)
#SC = O(n)    
```

```python
#238. Product of Array Except Self
class Solution(object):
    def productExceptSelf(self, nums):
        prefix = 1
        n = len(nums)
        results = n * [1]
        for i in range(len(nums)):
            results[i] = prefix
            prefix *= nums[i]
        
        suffix = 1
        for i in range(len(nums) -1 ,-1,-1):
            results[i] *= suffix
            suffix *= nums[i]
        
        return results
#TC - O(n)
#SC - O(1)
```

```python
#39. Combination Sum
class Solution(object):
    def combinationSum(self, candidates, target):
        def backtracking(start,remaining,combination):

            if remaining == 0:
                results.append(list(combination))
                return
            for i  in range(start,len(candidates)):
                if candidates[i] > remaining:
                    continue
                combination.append(candidates[i])
                backtracking(i,remaining - candidates[i],combination)
                combination.pop()
            
        results = []
        backtracking(0,target,[])
        return results
#TC - O(2^T)
#SC - O(T)
```

```python
#Rotate Array - 189
class Solution:
    def rotate(self, nums, k):
        n = len(nums)
        k = k%n
        temp = nums[-k:]
        for i in range(n-k-1,-1,-1):
            nums[k+i]= nums[i]
        for i in range(len(temp)):
            nums[i]= temp[i]
#TC = O(n)
#SC = O(k)
```

```python
#121. Best Time to Buy and Sell Stock
class Solution(object):
    def maxProfit(self, prices):

        buy = prices[0]
        profits = 0
        for sell in prices[1:]:
            if sell > buy:
                profits = max(profits,sell - buy)
            else:
                buy = sell

        return profits
    

def test_maxProfit():
    solution = Solution()

    # Test case: Prices list where we can make a profit
    prices = [7, 1, 5, 3, 6, 4]
    expected = 5  # The best profit is buying at 1 and selling at 6
    result = solution.maxProfit(prices)

    # Check if the result is as expected
    assert result == expected, f"Test failed: Expected {expected}, but got {result}"

    print("Test passed successfully!")

# Run the test script
test_maxProfit()

#Complexity
# Time  - O(n)
# Space - O(1)
```

```python
#1. Two Sum
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        seen = {}
        for i,num in enumerate(nums):
            diff = target - num
            if diff in seen:
                return (seen[diff],i)
            else:
                seen[num] = i
        

solution = Solution()
result = solution.twoSum([2, 7, 11, 15], 9)
print(result)  # Output: [0, 1]


#Complexity
# Time  - O(n)
# Space - O(n)
```

```python
#Anagram
def Anagram(s1,s2):
    s1 = s1.replace(" ","").lower()
    s2 = s2.replace(" ","").lower()
    
    if len(s1) != len(s2):
        return False
    
    count = {}

    for i in s1:
        if i in count:
            count[i] += 1
        else:
            count[i] = 1

    for i in s2:
        if i in count:
            count[i] -= 1
    else:
            count[i] = 1

    for k in count:
        if count[k] != 0:
            return False
```

```python
def Sum_Pair(Arr, K):
    seen = set()  # To keep track of the numbers we've seen
    output = set()  # To store unique pairs

    for i in Arr:
        target = K - i  # The number we need to find a pair with

        if target not in seen:  # Check if the target has been seen
            seen.add(i)  # Add the current number to the seen set
        else:
            output.add((min(i, target), max(i, target)))  # Add the pair to output

    return len(output)  # Return the number of unique pairs found
  
```

```python
# Find Missing Element
def Finder(arr1,arr2):
    arr1.sort()
    arr2.sort()

    for num1,num2 in zip(arr1,arr2):
        if num1 != num2:
            return num1
    return arr1[-1]
```

```python
#Largest Continous Sum
def largest_sum(arr):

    if not arr:  # Check for empty array
        return 0
    
    max_sum = cur_sum = arr[0]

    for num in arr[1:]:
        cur_sum = max(cur_sum + num, num)
        max_sum = max(max_sum,cur_sum)
    return max_sum
```

```python
#String Reversal
def reversal(s):
    words = []
    lenght = len(s)
    spaces = [' ']
    i = 0

    while i < lenght:

        if s[i] not in spaces:
            word_start = i

            while i< lenght and s[i] not in spaces:
                i +=1
            words.append(s[word_start:i])
        i +=1
    return " ".join(reversed(words))
```

```python
#Length of Last Word
class Solution(object):
    def lengthOfLastWord(self, s):
        count = 0
        for i in range(len(s) - 1,0,-1):
            if s[i] != " ":
                count +=1
            elif count > 0:
                break
            
        return count
# Time - O(n)
#Space - O(1)
```

```python
#Longest Common Prefix
class Solution(object):
    def longestCommonPrefix(self, strs):
        first_word = strs[0]
        for i in range(len(first_word)):
            for j in strs[1:]:
                if i>len(j) or first_word[i] != j[i]:
                    return first_word[0:i]  # Use slicing with a colon

#Time Complexity:O(S), where S is the sum of all characters in the input strings.
#Space Complexity - O(1)
```
